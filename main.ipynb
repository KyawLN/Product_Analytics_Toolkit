# PREDICTIVE PRIORITIZATION

```python
import pandas as pd
from textblob import TextBlob
import sys
import os

def load_data(filepath):
    """Safely loads CSV data."""
    if not os.path.exists(filepath):
        print(f"âŒ Error: File not found at {filepath}")
        sys.exit(1)
    return pd.read_csv(filepath)

def prioritize_features(df):
    """
    Calculates a 'Value Score' based on Revenue and User Requests vs Effort.
    This is a custom logic replacing subjective RICE scores.
    """
    print("\n--- ðŸš€ PROCESSING FEATURE PRIORITIZATION ---")
    
    # Normalize data for scoring (simple version)
    # Logic: (Revenue + (Requests * 10)) / Effort
    df['Value_Score'] = ((df['Projected_Revenue'] + (df['User_Requests_Count'] * 10)) / df['Dev_Effort_Days']).round(1)
    
    # Sort by highest score
    prioritized_df = df.sort_values(by='Value_Score', ascending=False).reset_index(drop=True)
    
    print("âœ… Top 3 Recommended Features to Build:")
    print(prioritized_df[['Feature_Name', 'Value_Score']].head(3))
    
    # Save report
    prioritized_df.to_csv('priority_report.csv', index=False)
    print("ðŸ’¾ Full report saved to 'priority_report.csv'")

def analyze_sentiment(df):
    """
    Analyzes user feedback using NLP to determine sentiment polarity.
    """
    print("\n--- ðŸ—£ï¸ PROCESSING SENTIMENT ANALYSIS ---")
    
    # Define a function to apply to each row
    def get_sentiment(text):
        blob = TextBlob(str(text))
        return round(blob.sentiment.polarity, 2)

    def get_label(score):
        if score > 0.3: return 'Positive'
        if score < -0.3: return 'Negative'
        return 'Neutral'

    # Apply the functions
    df['Sentiment_Score'] = df['Comment'].apply(get_sentiment)
    df['Sentiment_Label'] = df['Sentiment_Score'].apply(get_label)

# Calculate average sentiment
    avg_score = df['Sentiment_Score'].mean()
    print(f"ðŸ“Š Overall User Sentiment Score: {avg_score}")
    
    # Show critical issues (Negative sentiment)
    negatives = df[df['Sentiment_Label'] == 'Negative']
    print(f"âš ï¸ Found {len(negatives)} critical negative comments.")
    
    # Save report
    df.to_csv('sentiment_report.csv', index=False)
    print("ðŸ’¾ Full analysis saved to 'sentiment_report.csv'")

if __name__ == "__main__":
    print("ðŸ¤– Product Analytics Tool Initialized...")
    
    # 1. Load Data
    features_df = load_data('C:/Users/User/Documents/Docs/product_analytics_toolkit/data/features.csv')
    feedback_df = load_data('C:/Users/User/Documents/Docs/product_analytics_toolkit/data/feedback.csv')
    
    # 2. Run Analyses
    prioritize_features(features_df)
    analyze_sentiment(feedback_df)
    
    print("\nðŸŽ‰ Job Done!")
```
